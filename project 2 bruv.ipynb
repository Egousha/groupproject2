```python
#allows the user to input their message and what they want it to be shifted by
message=str(input('Enter your text:'))
shift=int(input('Enter your shift:'))
#function for the encryption,uses the two inputs
def encrypt(message, shift):

    #made the message a global variable, and an empty string
    global caesarmessage
    caesarmessage = ""

    #cycles through each character in the message
    for x in message:

        #for the uppercase characters
        if x.isupper(): 

            #gets a value between 0 and 25 for the letter
            x_ordval = ord(x) - ord('A')

            # shift the letter value by the shift value
            x_shiftpos = (x_ordval + shift) % 26 + ord('A')

            #converts the new letter value to a character
            x_newpos = chr(x_shiftpos)

            #adds the new letter to the string
            caesarmessage += x_newpos

        #for the lowercase characters
        elif x.islower():

            #gets a value  between 0 and 25 for the letter
            x_ordval = ord(x) - ord('a') 
            
            #shift the letter value by the shift value
            x_shiftpos = (x_ordval + shift) % 26 + ord('a')

            #converts the new letter value to a character
            x_newpos = chr(x_shiftpos)

            #adds the new letter to the string
            caesarmessage += x_newpos

        #for numbers
        elif x.isdigit():

            #shift its actual value 
            x_newpos = (int(x) + shift) % 10

            #adds new number to string
            caesarmessage += str(x_newpos)

        else:

            # if its a symbol, we couldn't get it to work because the gap and weird ascii values, so we left it, sorry.
            caesarmessage += x

    #returns the new completed encrypted message
    return caesarmessage

#the function for the decryption process
def decrypt(caesarmessage):
    #prints the encrypted message
    print('The encrypted message is', caesarmessage)
    
    #to figure out what the message was shifted by, possible shift values
    for key in range(1000):
  
        #make the decodedmessage global, and an empty string
        global decodedmessage
        decodedmessage = ""
 
        #goes through each character in the encrypted message
        for x in caesarmessage:

            #for uppercase values
            if x.isupper(): 

                #find the letter value between 0 and 25
                x_ordval = ord(x) - ord('A')

                # shift the letter value by whatever value the key is
                x_ogposit = (x_ordval - key) % 26 + ord('A')

                #converts the shifted letter value back to a string character
                x_orig = chr(x_ogposit)

                #adds the shifted letter value to the string
                decodedmessage += x_orig
           
            #for lowercase letters
            elif x.islower(): 

                #find the letter value between 0 and 25
                x_ordval = ord(x) - ord('a') 

                # shift the letter value by whatever value the key is
                x_ogposit = (x_ordval - key) % 26 + ord('a')

                #converts the shifted letter value back to a string character
                x_orig = chr(x_ogposit)

                #adds the shifted letter value to the string
                decodedmessage += x_orig
            elif x.isdigit():

                # if it's a number,shift its actual value 
                x_orig = (int(x) - key) % 10

                #adds the number to the string
                decodedmessage += str(x_orig)
            else:

                # if its neither alphabetical nor a number, just leave it like that
                decodedmessage += x
        #prints the possible shift value and what the decoded message would be for that value
        print('Possible Shift #%s: %s' % (key, decodedmessage))
        #breaks the loop once the decoded message is the same as the message the user entered
        if decodedmessage==message:
            break

#runs each function
encrypt(message, shift)
decrypt(caesarmessage)
```
